# Create a heat map with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 9)
```

```python
# Create random(ish) data.
rng = np.random.RandomState(42)
values = rng.randint(1, 1000, size=(6, 6))
data = pd.DataFrame(values, columns=list("ABCDEF"), index=list("GHIJKL"))
print(data)
```

```python
# Create the heat map.
plt.pcolormesh(data.values, cmap="Oranges", linewidth=1, edgecolor="white")
# Reverse the y axis so that the first row of the data frame is at the top of the chart.
plt.gca().invert_yaxis()
# Update the x and y tick labels.
plt.xticks(np.arange(6) + 0.5, data.columns, fontsize=16)
plt.yticks(np.arange(6) + 0.5, data.index, fontsize=16)
# Add labels for each data point.
for x in range(6):
    for y in range(6):
        plt.text(x + 0.5, y + 0.5, data.iloc[y, x], va="center", ha="center", fontsize=16)
# Add a colour bar.
plt.colorbar()

plt.show()
```
